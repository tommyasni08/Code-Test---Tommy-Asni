```python
import pandas as pd
import json
from datetime import datetime
import calendar
```

```python
# Question 2 Sensor Aggregation
with open("./JSON Files/sensor_data.json") as f:
    data = json.load(f)

# Question 3 Stream Sensor Aggregation
# with open("./SenStream/JSON Files/stream_sensor_data.json") as f:
#     data = json.load(f)
```

```python
df = pd.DataFrame(columns=['temperature','humidity','roomArea','id','timestamp'])
```

```python
for i in range(0,len(data['array'])) :
    df.loc[i] = [data['array'][i]['temperature'], data['array'][i]['humidity'],data['array'][i]['roomArea'], data['array'][i]['id'],data['array'][i]['timestamp']]
```

```python
df
```

```python
#  Create new columns 'day' and 'weekday'
df['date'] = df['timestamp']
df['day'] = df['timestamp']
df['weekday'] = df['timestamp']
```

```python
# Convert timestmap to datetime, day, and weekday 
for i in range(0,len(data['array'])) :
    df['date'][i] = datetime.fromtimestamp(df['timestamp'][i]/1e3)
    df['day'][i] = df['date'][i].day
    df['weekday'][i] = df['date'][i].weekday()
```

```python
df
```

```python
# Create function for Switch-Case Replacement Function to Convert the Weekday 
def weekdayConv(weekday) :
    if weekday == 6:
        return 'Sunday'
    elif weekday == 0:
        return 'Monday'
    elif weekday == 1:
        return 'Tuesday'
    elif weekday == 2:
        return 'Wednesday'
    elif weekday == 3:
        return 'Thursday'
    elif weekday == 4:
        return 'Friday'
    elif weekday == 5:
        return 'Saturday'

# Converting the weekday(number) column to weekday(string)
for i in range(0,len(data['array'])) :
    container = weekdayConv(df['weekday'][i])
    df['weekday'][i] = container
```
